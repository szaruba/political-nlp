```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
measures = pd.read_csv('../protocols/labelled/massnahmen_relevant.csv', sep='\t')
```

```python
speeches = measures['Speech']
speech_lenghts = speeches.map(lambda s: len(s.split(' '))).tolist()
speech_lenghts.sort()
```

```python
plt.hist(speech_lenghts, 25, density=False, histtype='barstacked', rwidth=0.8, facecolor='g',alpha=0.8, )
#plt.plot(speech_lenghts)
plt.title('Speech counts per length')
plt.xlabel('Words in speech')
plt.ylabel('Speech count')
plt.savefig('../results/measures_speech_lengths.png')
plt.show()
```
